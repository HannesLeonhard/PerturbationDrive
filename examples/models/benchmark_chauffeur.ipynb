### Chauffeur Baseline

```python
from perturbationdrive import Scenario

try:
    chauffeur.name = "Chauffeur"
    benchmark = PerturbationDrive(simulator=simulator_sdsandbox, ads=chauffeur)
    benchmark.simulator.connect()
    starting_pos = benchmark.simulator.initial_pos

    # init all scenarios
    scenarios = []
    counter = 0
    # iterate over all agents
    for angle, segment in zip(angles, segments):
        # run all perturbations for each road
        waypoints = road_gen.generate(
            starting_pos=starting_pos, angles=angle, seg_length=segment
        )
        scenarios = [
            Scenario(
                waypoints=waypoints,
                perturbation_function="",
                perturbation_scale=0,
            )
        ]
        print(f"Starting Agent {chauffeur.name} on road {angle}")
        # run the benchmark
        res = benchmark.simulate_scenarios(
            scenarios=scenarios,
            log_dir=f"./logs/donkey_baseline_perturbations_{chauffeur.name}.json",
        )
        counter += 1
        print(f"Finished Agent {chauffeur.name} on road {angle}")
except Exception as e:
    print(e)
finally:
    benchmark.simulator.tear_down()
```

#### Donkey Chauffeur RQ2

```python
from perturbationdrive import Scenario
import gc

try:
    chauffeur.name = "Chauffeur"
    benchmark = PerturbationDrive(simulator=simulator_sdsandbox, ads=chauffeur)
    benchmark.simulator.connect()
    starting_pos = benchmark.simulator.initial_pos

    # init all scenarios
    scenarios = []
    counter = 0
    # iterate over all agents
    for angle, segment in zip(angles, segments):
        # run all perturbations for each road
        waypoints = road_gen.generate(
            starting_pos=starting_pos, angles=angle, seg_length=segment
        )
        scenarios = []
        for func in perturbations:
            for i in range(5):
                scenarios.append(
                    Scenario(
                        waypoints=waypoints,
                        perturbation_function=func,
                        perturbation_scale=i,
                    )
                )
        print(f"Starting Agent {chauffeur.name} on road {angle}")
        # run the benchmark
        res = benchmark.simulate_scenarios(
            scenarios=scenarios,
            log_dir=f"./logs/donkey_rq2_perturbations_road{counter}_{chauffeur.name}.json",
        )
        del res
        del scenarios
        # collect garbage
        gc.collect()
        counter += 1
        print(f"Finished Agent {chauffeur.name} on road {angle}")
except Exception as e:
    print(e)
finally:
    benchmark.simulator.tear_down()
```
