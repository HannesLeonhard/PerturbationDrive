#### Simulate all scenarios

```python
from perturbationdrive import PerturbationDrive
from examples.models.dave2_agent import Dave2Agent
from examples.models.epoch_agent import EpochAgent
from examples.models.chauffeur_agent import ChauffeurAgent

dave2 = Dave2Agent()
epoch = EpochAgent()
chauffeur = ChauffeurAgent()
```

```python
agents = [dave2, epoch, chauffeur]

perturbations = [
    "gaussian_noise",
    "poisson_noise",
    "impulse_noise",
    "defocus_blur",
    "glass_blur",
    "motion_blur",
    "zoom_blur",
    "increase_brightness",
    "contrast",
    "elastic",
    "pixelate",
    "jpeg_filter",
    "shear_image",
    "translate_image",
    "scale_image",
    "rotate_image",
    "fog_mapping",
    "splatter_mapping",
    "dotted_lines_mapping",
    "zigzag_mapping",
    "canny_edges_mapping",
    "speckle_noise_filter",
    "false_color_filter",
    "high_pass_filter",
    "low_pass_filter",
    "phase_scrambling",
    "histogram_equalisation",
    "reflection_filter",
    "white_balance_filter",
    "sharpen_filter",
    "grayscale_filter",
    "fog_filter",
    "frost_filter",
    "snow_filter",
    "posterize_filter",
    "cutout_filter",
    "sample_pairing_filter",
    "gaussian_blur",
    "saturation_filter",
    "saturation_decrease_filter",
]
```

```python
from perturbationdrive import Scenario

try:
    benchmark = PerturbationDrive(simulator=simulator_sdsandbox, ads=agents[0])
    benchmark.simulator.connect()
    starting_pos = benchmark.simulator.initial_pos

    # init all scenarios
    scenarios = []
    for angle, segment in zip(angles, segments):
        waypoints = road_gen.generate(
            starting_pos=starting_pos, angles=angle, seg_length=segment
        )
        for func in perturbations:
            for i in range(5):
                scenarios.append(
                    Scenario(
                        waypoints=waypoints,
                        perturbation_function=func,
                        perturbation_scale=i,
                    )
                )

    for agent in agents:
        benchmark.setADS(agent)
        # run the benchmark
        res = benchmark.simulate_scenarios(
            scenarios=scenarios, log_dir=f"./logs/donkey_benchmark_normal_perturbations_{agent.name}.json"
        )
except Exception as e:
    print(e)
finally:
    benchmark.simulator.tear_down()
```

```python
try:
    benchmark.setSimulator(simulator_udacity)
    benchmark.simulator.connect()
    starting_pos = benchmark.simulator.initial_pos

    # init all scenarios
    scenarios = []
    for angle, segment in zip(angles, segments):
        waypoints = road_gen.generate(
            starting_pos=starting_pos, angles=angle, seg_length=segment
        )
        scenarios.append(
            Scenario(
                waypoints=waypoints,
                perturbation_function="",
                perturbation_scale=0,
            )
        )

    for agent in agents:
        benchmark.setADS(agent)
        # run the benchmark
        res = benchmark.simulate_scenarios(
            scenarios=scenarios, log_dir=f"./logs/udacity_benchmark_normal_perturbations_{agent.name}.json"
        )

except Exception as e:
    print(e)
finally:
    benchmark.simulator.tear_down()
```
