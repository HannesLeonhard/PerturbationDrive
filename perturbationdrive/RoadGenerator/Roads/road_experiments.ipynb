```python
from perturbationdrive.RoadGenerator.Roads.road import Road
from typing import List, Tuple
import matplotlib.pyplot as plt


def visualize_simulator_raod_grid(roads: List[Road]):
    # calculate the amount of turns for each road
    amount_turns: List[Tuple[int, float]] = [
        road.compute_num_turns()[0] for road in roads
    ]
    smoothness: List[float] = [road.compute_curvature() for road in roads]

    # create bins fpr amount of turns and smoothness
    bins_turns: List[int] = [x for x in range(min(amount_turns), max(amount_turns) + 1)]

    # create n bins between the min and max smoothness
    range_smoothnes = max(smoothness) - min(smoothness)
    interval_smoothnes = range_smoothnes / len(roads)
    counter = min(smoothness)
    bins_smoothness = [counter]
    while counter < max(smoothness):
        counter += interval_smoothnes
        bins_smoothness.append(counter)

    # create a collection of plots with the amount of turns on the x-axis and the smoothness on the y-axis
    fig, axs = plt.subplots(len(bins_turns), len(bins_smoothness))

    # place each road at the correct position in the collection of plots
    for road in roads:
        # find index in the bins_turns list
        index_turns = bins_turns.index(road.compute_num_turns()[0])
        # find index in the bins_smoothness list
        index_smoothness = 0
        for i in range(len(bins_smoothness)):
            if bins_smoothness[i] > road.compute_curvature():
                index_smoothness = i - 1
                break
        print(f"index turns: {index_turns}; index smoothness: {index_smoothness}")
        # place the road in the correct plot
        axs[index_turns, index_smoothness].plot(
            road.get_concrete_representation(to_plot=True)[0],
            road.get_concrete_representation(to_plot=True)[1],
            color="red",
            linewidth=2,
            linestyle="-",
        )
        print(f"Turns: {road.compute_num_turns()[0]}")
        print(f"Smoothness: {road.calculate_smoothness()}")
        print(f"Curvature: {road.compute_curvature(w=4)}")

    print(f"Bins are {bins_turns} and {bins_smoothness}")
```

```python
import numpy as np

curvature_min = np.inf
curvature_max = -np.inf
for _ in range(100):
    angles = [randint(-65, 65) for _ in range(8)]
    seg_length = [randint(15, 35) for _ in range(8)]
    _ = road_generator.generate(
        starting_pos=(0, 0, 0, 4),
        angles=angles,
        seg_length=seg_length,
    )
    if road_generator.previous_road.compute_curvature(w=4) < curvature_min:
        curvature_min = road_generator.previous_road.compute_curvature(w=4)
    if road_generator.previous_road.compute_curvature(w=4) > curvature_max:
        curvature_max = road_generator.previous_road.compute_curvature(w=4)

print(f"curvature min: {curvature_min}, max: {curvature_max}")
```

```python
range_curvature = curvature_max - curvature_min
interval_curvature = range_curvature / 3
curvature_bins = [
    curvature_min,
    curvature_min + interval_curvature,
    curvature_min + 2 * interval_curvature,
    curvature_max,
]
turn_bins = [1, 2, 3]

roads = [[None, None, None], [None, None, None], [None, None, None]]
angles_list = [[None, None, None], [None, None, None], [None, None, None]]
seg_lengths_list = [[None, None, None], [None, None, None], [None, None, None]]
# we generate roads until we find roads that are in the correct bin
loop = True

while loop:
    found_all = True
    for i in range(3):
        for j in range(3):
            if roads[i][j] is None:
                found_all = False
                break
    if found_all:
        loop = False
        break

    angles = [randint(-65, 65) for _ in range(8)]
    angles[0] = 0
    seg_length = [randint(15, 35) for _ in range(8)]
    _ = road_generator.generate(
        starting_pos=(0, 0, 0, 4),
        angles=angles,
        seg_length=seg_length,
    )
    curvature = road_generator.previous_road.compute_curvature(w=4)
    turns = road_generator.previous_road.compute_num_turns()[0]

    # find the correct bin for the road
    if curvature <= curvature_bins[1]:
        curvature_bin = 0
    elif curvature <= curvature_bins[2]:
        curvature_bin = 1
    else:
        curvature_bin = 2

    if turns == turn_bins[0]:
        turn_bin = 0
    elif turns == turn_bins[1]:
        turn_bin = 1
    elif turns == turn_bins[2]:
        turn_bin = 2
    else:
        continue
    if roads[curvature_bin][turn_bin] is None:
        roads[curvature_bin][turn_bin] = road_generator.previous_road
        angles_list[curvature_bin][turn_bin] = angles
        seg_lengths_list[curvature_bin][turn_bin] = seg_length
        print(f"curvature: {curvature}, turns: {turns}")

```

```python
# print angles and seg_length, smppthnes, turns and visualize for each road
for i in range(3):
    for j in range(3):
        print(f"road {i}, {j}")
        print(f"angles: {angles_list[i][j]}")
        print(f"seg_length: {seg_lengths_list[i][j]}")
        print(f"smoothness: {roads[i][j].calculate_smoothness()}")
        print(f"turns: {roads[i][j].compute_num_turns()[0]}")
        visualize_road(roads[i][j])
```
